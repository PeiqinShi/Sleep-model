```python
data = pd.read_csv('GABAA.csv',encoding='gb18030')
```

### ECFP4

```python
data['smiles']
```

```python
featurizer = dc.feat.CircularFingerprint(size=1024,radius=4)
agonist_features = featurizer.featurize(data['smiles'][0:244])
antagonist_features = featurizer.featurize(data['smiles'][244:])
```

```python
data_ecfp = np.concatenate((agonist_features, antagonist_features), axis=0)

# è¿è¡Œt-SNEç®—æ³•
tsne = TSNE(n_components=2)
transformed_data = tsne.fit_transform(data_ecfp)

# ç»˜åˆ¶ç»“æžœ
sns.scatterplot(x=transformed_data[:, 0], y=transformed_data[:, 1], hue=['Positive']*244 + ['Negative']*244)
```

# Similarity

```python
import pandas as pd
from rdkit import Chem
from rdkit.DataStructs import FingerprintSimilarity
```

```python
data = pd.read_csv('GABAA.csv',encoding='gb18030')
plant = pd.read_csv('plant_data.csv',encoding='gb18030')
```

```python
plant
```

```python
# å®šä¹‰æ•°æ®é›†1çš„SMILES
dataset1_smiles = data['smiles']

# å®šä¹‰æ•°æ®é›†2çš„SMILES
dataset2_smiles = plant['smiles']

# åˆ›å»ºç»“æžœDataFrame
AD_results = pd.DataFrame(columns=['SMILES', 'InDomain'])

# è®¡ç®—æ•°æ®é›†2ä¸­æ¯ä¸ªåˆ†å­ä¸Žæ•°æ®é›†1çš„ç›¸ä¼¼åº¦ï¼Œå¹¶åˆ¤æ–­æ˜¯å¦åœ¨åº”ç”¨åŸŸå†…
for smiles in dataset2_smiles:
    # è½¬æ¢ä¸ºRDKitåˆ†å­å¯¹è±¡
    mol2 = Chem.MolFromSmiles(smiles)
    
    # åˆå§‹åŒ–åº”ç”¨åŸŸæ ‡å¿—
    in_domain = False
    
    # è®¡ç®—ä¸Žæ•°æ®é›†1ä¸­æ¯ä¸ªåˆ†å­çš„ç›¸ä¼¼åº¦
    for ref_smiles in dataset1_smiles:
        # è½¬æ¢ä¸ºRDKitåˆ†å­å¯¹è±¡
        mol1 = Chem.MolFromSmiles(ref_smiles)
        
        # è®¡ç®—Tanimotoç›¸ä¼¼åº¦
        similarity = FingerprintSimilarity(Chem.RDKFingerprint(mol1), Chem.RDKFingerprint(mol2))
        #print(f"Tanimoto similarity between {smiles} and {ref_smiles}: {similarity}")
        # åˆ¤æ–­æ˜¯å¦åœ¨åº”ç”¨åŸŸå†…
        if similarity > 0.7: 
            in_domain = True
    
    # æ·»åŠ ç»“æžœåˆ°DataFrame
    AD_results = AD_results.append({'SMILES': smiles, 'InDomain': in_domain}, ignore_index=True)
```

```python
AD_results
```

```python
# è®¡ç®—Trueå’ŒFalseå€¼çš„æ•°é‡
value_counts = AD_results['InDomain'].value_counts()

# æ‰“å°ç»“æžœ
print(value_counts)
```
