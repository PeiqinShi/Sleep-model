```python
import numpy as np
import pandas as pd
import deepchem as dc
import sklearn
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.model_selection import KFold
from sklearn import metrics
from sklearn.metrics import confusion_matrix, roc_auc_score, accuracy_score, precision_score, recall_score
from sklearn.feature_selection import VarianceThreshold
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce
import time
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/GABAA.csv', encoding='gb18030')
```

```python
def featurizer(featname, data):
    if featname == "MACCS":
        featurizer = dc.feat.MACCSKeysFingerprint()
    elif featname == "RDkit":
        featurizer = dc.feat.RDKitDescriptors()
    else:
        raise ValueError("Unknown feature type")

    features = featurizer.featurize(data['smiles'])
    vt = VarianceThreshold(threshold=(.98 * (1 - .98)))
    X_new = vt.fit_transform(features)
    
    dataset = dc.data.NumpyDataset(X_new, data['class'])
    splitter = dc.splits.RandomSplitter()
    train_dataset, test_dataset = splitter.train_test_split(dataset=dataset, frac_train=0.8, seed=200)

    return train_dataset.X, test_dataset.X, train_dataset.y, test_dataset.y
```

```python
# 选择模型并进行网格搜索
def train_model(modelname, X_train, y_train):
    if modelname == "rf_maccs":
        model = RandomForestClassifier(random_state=42)
        param_grid = {'n_estimators':[80,100,120,140,160,180,200,220],
              'max_depth':[5,8,10,20,30,40],
            'max_features':["auto", "sqrt", "log2"]}
        
    elif modelname == "rf_rdkit":
        model = RandomForestClassifier(random_state=42)
        param_grid = {'n_estimators':[140,160,180,200,220,240,260,280,300],
              'max_depth':[5,8,10,20,30,40],
            'max_features':["auto", "sqrt", "log2"]}
        
    elif modelname == "svm_maccs":
        model = SVC(probability=True, random_state=42)
        param_grid = {'C': [0.001, 0.005, 0.01, 0.05, 0.1, 0.25, 0.5, 0.75, 1, 2.5, 5, 7.5, 10, 25, 50, 75, 100], 
              'kernel': ['linear', 'rbf','poly'],
             'gamma':[0.001, 0.005, 0.01, 0.05, 0.1, 0.25, 0.5, 0.6, 0.7, 0.8, 0.8, 0.9, 1]
             }
        
    elif modelname == "xgb_maccs":
        model = XGBClassifier(random_state=42)
        param_grid = {'n_estimators':[80,100,120,140,160,180],
                  'max_depth':[3,5,7,9,11,13],
                  'colsample_bytree':[0.1,0.3,0.5,0.7,0.9],
                 'learning_rate':[0.01,0.05, 0.1,0.15,0.2]
                 }
        
    else:
        raise ValueError("Unknown model name")

    # 进行网格搜索
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='roc_auc', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    
    print(f"Best parameters for {modelname}: {grid_search.best_params_}")
    return grid_search.best_estimator_
```

```python
def SelectModel(modelname):
    if modelname == "rf_maccs":
        return RandomForestClassifier(max_depth=8,max_features="log2", n_estimators=180, random_state=100)
    elif modelname == "rf_rdkit":
        return RandomForestClassifier(max_depth=20,  max_features="log2",n_estimators=260, random_state=100)
    elif modelname == "svm_maccs":
        return SVC(probability=True, C=1, gamma=0.1, kernel="rbf", random_state=100)
    elif modelname == "xgb_maccs":
        return XGBClassifier(colsample_bytree=0.1, learning_rate=0.1, max_depth=9, n_estimators=140, random_state=100)
    else:
        raise ValueError("Unknown model name")
```

```python
def SelectModel(modelname):
    if modelname == "rf_maccs":
        return RandomForestClassifier(n_estimators=100, random_state=42)
    elif modelname == "rf_rdkit":
        return RandomForestClassifier(n_estimators=100, random_state=42)
    elif modelname == "svm_maccs":
        return SVC(probability=True, random_state=42)
    elif modelname == "xgb_maccs":
        return XGBClassifier(random_state=42)
    else:
        raise ValueError("Unknown model name")
```

```python
def get_oof(clf, n_folds, X_train, y_train, X_test):
    ntrain = X_train.shape[0]
    ntest = X_test.shape[0]
    classnum = len(np.unique(y_train))
    kf = KFold(n_splits=n_folds, random_state=42, shuffle=True)

    oof_train = np.zeros((ntrain, classnum))
    oof_test = np.zeros((ntest, classnum))

    for train_index, test_index in kf.split(X_train):
        kf_X_train, kf_y_train = X_train[train_index], y_train[train_index]
        kf_X_test = X_train[test_index]

        clf.fit(kf_X_train, kf_y_train)
        oof_train[test_index] = clf.predict_proba(kf_X_test)
        oof_test += clf.predict_proba(X_test)

    oof_test /= float(n_folds)
    return oof_train, oof_test
```

```python
modelist = ['rf_maccs','rf_rdkit','svm_maccs','xgb_maccs']
newfeature_list = []
newtestdata_list = []

# 初始化进度条
progress_bar = tqdm(modelist, desc="Training Models", unit="model")

for modelname in progress_bar:
    start_time = time.time()  # 记录开始时间
    clf_first = SelectModel(modelname)
    
    if modelname == 'rf_maccs':
        data_train, data_test, label_train, label_test = featurizer('MACCS', data)
    elif modelname == 'rf_rdkit':
        data_train, data_test, label_train, label_test = featurizer('RDkit', data)
    elif modelname == 'svm_maccs':
        data_train, data_test, label_train, label_test = featurizer('MACCS', data)
    elif modelname == 'xgb_maccs':
        data_train, data_test, label_train, label_test = featurizer('MACCS', data)
    else:
        raise ValueError("Unknown model name")

    # 重新训练基础模型
    #clf_first = train_model(modelname, data_train, label_train)
    
    # 进行stacking
    oof_train_, oof_test_ = get_oof(clf=clf_first, n_folds=5, X_train=data_train, y_train=label_train, X_test=data_test)
    newfeature_list.append(oof_train_)
    newtestdata_list.append(oof_test_)

    end_time = time.time()  # 记录结束时间
    elapsed_time = end_time - start_time  # 计算耗时

    # 更新进度条描述信息
    progress_bar.set_description(f"Training {modelname} ({elapsed_time:.2f}s)")

    # 也可以使用 tqdm.write() 单独输出时间
    tqdm.write(f"Model {modelname} training and stacking completed in {elapsed_time:.2f} seconds.")
```

```python
# 特征组合
newfeature = reduce(lambda x, y: np.concatenate((x, y), axis=1), newfeature_list)
newtestdata = reduce(lambda x, y: np.concatenate((x, y), axis=1), newtestdata_list)
```

```python
# 第二级模型
clf_second = RandomForestClassifier()
clf_second.fit(newfeature, label_train)
```

# seed=100

```python
# 预测
pred_proba = clf_second.predict_proba(newtestdata)
y_test_pred = clf_second.predict(newtestdata)
```

```python
confusion_mat_1 = confusion_matrix(label_test, y_test_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat_1, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")
plt.title("Confusion Matrix")
plt.show()
```

```python
test_pred_list = []
for test_score in pred_proba:
    test_score = test_score[1]
    test_pred_list.append(test_score)
```

```python
test_pred_array = np.array(test_pred_list)
```

```python
auc = roc_auc_score(label_test, test_pred_array)
acc = accuracy_score(label_test, np.round(test_pred_array))
pre = precision_score(label_test, np.round(test_pred_array))
rec = recall_score(label_test, np.round(test_pred_array))

print(auc,acc,pre,rec)
```
