```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Family

### Model prediction score

```python
df = pd.read_excel('../data/histogram.xlsx',sheet_name='Family')

# Sort the DataFrame by the 'Pred' column in descending order and select the top 20 rows
df_top20 = df.sort_values(by='Pred', ascending=False).head(20)

# Set the default font family to 'Arial' for plotting
plt.rcParams['font.family'] = 'Arial'

# Create a figure and axis object with a specified size of 10x24 inches
fig, ax =plt.subplots(figsize=(10, 24))
# Define a colormap (color gradient) using the 'Greens' palette from Matplotlib
cmap = plt.cm.Greens
# Normalize the 'Pred' values to a range between 0 and 1 for color mapping
norm = plt.Normalize(df_top20['Pred'].min(),df_top20['Pred'].max())
colors = cmap(norm(df_top20['Pred']))
# Create a bar plot using Seaborn, with 'Pred' on the x-axis and 'Family' on the y-axis
# The color of each bar is determined by the normalized 'Pred' values
sns.barplot(x="Pred", y="Family", data=df_top20, palette=colors, ax=ax)
# Create a ScalarMappable object for the colorbar using the colormap and normalization
sm = plt.cm.ScalarMappable(cmap=cmap,norm=norm)
sm.set_array([])
# Add a colorbar to the plot
cbar = fig.colorbar(sm,
                    ax=ax, 
                    label="Model prediction score",
                    orientation="vertical",
                    aspect=100,
                    fraction=0.02
                   )
# Set the label for the colorbar with a specified font size
cbar.set_label("Model prediction score", fontsize=24) 
# Set the x-axis label with a specified font size
ax.set_xlabel("Model prediction score",fontsize=24)
# Set the y-axis label with a specified font size
ax.set_ylabel("Family name",fontsize=40)
# Set the font size for y-axis tick labels
ax.set_yticklabels(ax.get_yticklabels(), fontsize=24)

# Adjust the layout to fit everything neatly
plt.tight_layout()

#plt.savefig('plant_family_probability.png', dpi=1200)
plt.show()
```

### Relative content of sleep-promoting compounds

```python
df = pd.read_excel('../data/histogram.xlsx',sheet_name='Family')

# Sort the DataFrame by the 'Content' column in descending order and select the top 20 rows
# This step ensures that we get the families with the highest content values
df_top20 = df.sort_values(by='Content', ascending=False).head(20)
plt.rcParams['font.family'] = 'Arial'

# Create a figure and axis object with a specified size of 10x24 inches
# This size is chosen to accommodate the vertical bar plot with many categories
fig, ax =plt.subplots(figsize=(10, 24))

# Define a colormap (color gradient) using the 'Greens' palette from Matplotlib
# The color gradient will be used to represent the 'Content' values in the bar plot
cmap = plt.cm.Greens

# Normalize the 'Content' values to a range between 0 and 1 for color mapping
# This step ensures that the color gradient is applied proportionally to the values
norm = plt.Normalize(df_top20['Content'].min(),df_top20['Content'].max())
colors = cmap(norm(df_top20['Content']))

# Create a bar plot using Seaborn, with 'Content' on the x-axis and 'Family' on the y-axis
# The color of each bar is determined by the normalized 'Content' values
sns.barplot(x="Content", y="Family", data=df_top20, palette=colors, ax=ax)

# Create a ScalarMappable object for the colorbar using the colormap and normalization
# This object is necessary to create a colorbar that represents the color gradient
sm = plt.cm.ScalarMappable(cmap=cmap,norm=norm)
sm.set_array([])

# Add a colorbar to the plot
# The colorbar provides a visual representation of the 'Content' values corresponding to the colors
cbar = fig.colorbar(sm,
                    ax=ax, 
                    label="Relative content of sleep-promoting compounds",
                    orientation="vertical",
                    aspect=100,
                    fraction=0.02
                   )

# Set the label for the colorbar with a specified font size
cbar.set_label("Relative content of sleep-promoting compounds", fontsize=24)
# Set the x-axis label with a specified font size
ax.set_xlabel("Relative content of sleep-promoting compounds",fontsize=24)
# Set the y-axis label with a specified font size
ax.set_ylabel("Family name",fontsize=40)
# Set the font size for y-axis tick labels to ensure readability
ax.set_yticklabels(ax.get_yticklabels(), fontsize=24)
plt.tight_layout()
#plt.savefig('plant_family_probability_content.png', dpi=1200)
plt.show()
```
